```python
1. Docstrings (Documentation Strings)
A Docstring is a special string used to document what a specific function, class, or module does. Unlike regular comments (#), docstrings are stored as an attribute of the object, meaning they can be accessed while the code is running.

Syntax: Written inside triple quotes ("""...""").

Location: Must be the very first line inside a function or class.
```

```python
def add_numbers(a, b):
    """Returns the sum of two integers."""
    return a + b

# You can actually print the docstring like this:
print(add_numbers.__doc__) 
# Output: Returns the sum of two integers.
```

```python
2. PEP 8 (Python Enhancement Proposal 8)
PEP 8 is the official Style Guide for Python Code. It is a set of rules that tells you how to format your code so it is clean, consistent, and easy for other humans to read.

Key PEP 8 Rules:
Indentation: Use 4 spaces per indent level (not tabs).

Naming: Use snake_case for functions and variables; PascalCase for classes.

Spacing: Use one space around operators (e.g., x = 5, not x=5).

Line Length: Limit all lines to a maximum of 79 characters.

```
