```python
# 1. Take input from the user for three numbers on a single line.
#    The numbers are expected to be separated by spaces.
user_input_string = input("Enter three numbers, separated by spaces (e.g., 10 5 8): ")

# 2. Split the input string into a list of strings.
#    For example, "10 5 8" becomes ['10', '5', '8'].
numbers_as_strings = user_input_string.split()

# 3. Convert each string in the list to an integer.
#    map(int, ...) applies the int() function to every item in numbers_as_strings.
#    list(...) converts the map object into a proper list of integers.
#    So, ['10', '5', '8'] becomes [10, 5, 8].
numbers_as_integers = list(map(int, numbers_as_strings))

# 4. Find the smallest number in the list using the built-in min() function.
#    min() takes a list of numbers and returns the smallest one.
smallest_number = min(numbers_as_integers)

# 5. Print the original numbers entered and the smallest number found.
print(f"The numbers you entered are: {numbers_as_integers}")
print(f"The smallest number is: {smallest_number}")
```
