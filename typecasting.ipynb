```python
# definition of typecasting
# Typecasting in programming refers to the process of converting a variable from one data type to another. 
# This is often necessary when you want to perform operations that require a specific data type. 
# For example, converting a string to an integer so you can perform arithmetic operations on it.

# In Python, you can use functions like int(), float(), and str() to convert between different types.
```

```python
no = 5
n1 = 6
print (str(n1)+str(no))
```

```python
no = 5
n1 = 6
print ((n1)+(no))
```
