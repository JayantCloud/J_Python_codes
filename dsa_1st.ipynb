```python
def gcd(m, n):
    # Step 1: Find factors of m
    fm = []
    for i in range(1, m + 1):
        if (m % i) == 0:
            fm.append(i)

    # Step 2: Find factors of n
    fn = []
    for j in range(1, n + 1):
        if (n % j) == 0:
            fn.append(j)
            
    # Step 3: Find common factors
    cf = []
    for f in fm:
        if f in fn:
            cf.append(f)
            
    # Step 4: Return the largest common factor
    return cf[-1]

# How to use it:
num1 = int(input('Give the first number: '))
num2 = int(input('Give the second number: '))
print("The GCD is:", gcd(num1, num2))
```
