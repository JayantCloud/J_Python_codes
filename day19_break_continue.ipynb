the break statement enables the program to skip over the part of code i.e the break statement terminates the very loop it lies within

```python
for i in range(10):
    print(" 5 times",i+1 , "=", 5 * (i+1))
    if(i == 5):
        break


    
```

# continue statement
the continue statement skips the rest of loop statement and causes the next iteration to occur.

```python
for i in range(10):
    if(i==6):
        print("skipped the iteration")
        continue
    print(" 6 times", i+1, 6*(i+1))
```
