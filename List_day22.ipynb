```python
m = ["a", 12.4, 43,23,56,785,123,'k']
n = m.copy()
print(n)
```

now there is another way to use this copy function i.e. 
    assigning the value directly into the 
new variable, but if you will change the list values then 
what the original list will also change.
    

```python
j = [12,23,42,87,456,90,212]
s = j #here the value has been assigned in new variable
j.append(11)
print(j, s)
```

now as you can see above changing j changes the s also even the j has been assigned to the value of s
hence we use copy function which will copy this list and the original list will remain intact.
    

```python
j = [12,23,42,87,456,90,212]
j.extend(m)
print(j)
```

```python
here  the values of list j has been kept behind the list m
```
