What are Strings?
A string is simply a sequence of characters (letters, numbers, symbols, or spaces) wrapped in quotes. In Python, you can use either single quotes ('...') or double quotes ("...").

What is String Formatting?
String formatting (also called string interpolation) is the process of building a dynamic string by inserting variables or expressions
into a piece of predefined text.

Three Ways to Format Strings
Python has evolved over the years, giving us three main ways to do this.

1. F-Strings (The Modern Way)
Introduced in Python 3.6, these are the fastest and easiest to read. 
    You just put an f before the opening quote and use {} for your variables.

```python
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")
# Output: My name is Alice and I am 25 years old.
```

2. The .format() Method (The "New" Style)
Used frequently before f-strings, this uses curly braces as placeholders and calls a method at the end.

```python
print("My name is {} and I am {}".format("Bob", 30))
```

```python
3. The % Operator (The "Old" Style)
This is the oldest method (inherited from the C language). You'll still see it in older codebases.
```

```python
print("Hello, %s" % "Charlie") 
# %s stands for 'string'
```
