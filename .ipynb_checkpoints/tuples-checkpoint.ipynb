In Python, a tuple is an ordered collection of items that is immutable, meaning it cannot be changed once it is created.

Think of it like a "locked" list. You define a tuple using parentheses () instead of square brackets.

```python
t =(1,2)
print(t)
print(type(t))
```

```python
**tuples methods**
Feature,List Methods ([]),Tuple Methods (())
Search,"index(), count()","index(), count()"
Add,"append(), extend(), insert()",None
Remove,"pop(), remove(), clear()",None
Modify,"sort(), reverse()",None 

  in Tuple as we know we cant change the value of tuple
  so we have to change it in list first then in tuple
  however while concatenating we can concatenate two tuples together
  without changing it into the list.
  
```

```python
country = ("spain","russia","morroco","canada")
t = list(country)
print(t)
```

```python
# now we can change in the tuples
t.append("russia")
print(t)
t.pop(2)

# pop(index): Removes and returns the item at the given index. If you don't provide an index, it 
# removes the last item.
```
