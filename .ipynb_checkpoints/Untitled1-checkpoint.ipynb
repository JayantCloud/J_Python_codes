```python
i = 0
while(i < 3):
   print(i)  # e.g., 3 spaces
   i = i+1   # e.g., 4 spaces or a tab --> IndentationError
```

```python
# recursion
#  Print the numbers from 1 to 10 recursively
def factorial(n):
    if n == 0: 
        return 1
    return n * factorial(n -  1)
```

```python
def numb(n):
    if n == 0:
        return
    numb(n-1)
    print(n)
numb(10)
```

```python
def print_reverse(n):
    if n == 0:   # base case
        return
    print(n)          # print first
    print_reverse(n-1)  # then recurse down

print_reverse(10)
```
