**Keywrord Arguments**


Keyword arguments in Python are a way to pass arguments to a function by explicitly specifying the parameter names along with their values. This makes the code more readable and allows you to pass arguments in any order, as long as the parameter names are correctly specified.




```python
def greet(name, age, city):
    print(f"Hello, {name}! You are {age} years old and live in {city}.")

# Using keyword arguments
greet(age=25, city="Delhi", name="Amit")
```

```python
def get_ordinal(n):
    # Handles 1st, 2nd, 3rd, ..., 21st, 22nd, etc.
    if 10 <= n % 100 <= 20:
        suffix = 'th'
    else:
        suffix = {1: 'st', 2: 'nd', 3: 'rd'}.get(n % 10, 'th')
    return f"{n}{suffix}"

total_numbers = int(input("How many numbers do you have? "))
numbers = []

for i in range(1, total_numbers + 1):
    prompt = f"Give {get_ordinal(i)} number: "
    num = input(prompt)
    numbers.append(num)

print("Numbers entered:", numbers)
```

1. Arbitrary Arguments (**args**)
When you prefix a parameter with a single asterisk (*), the function receives a tuple of arguments.

Use Case: When you want to perform the same operation on a list of items (like summing numbers or concatenating strings).

The Name: You can technically name it anything (e.g., *elements), but *args* is the standard convention.

2. Arbitrary Keyword Arguments (**kwargs**)
When you prefix a parameter with two asterisks (**), the function receives a dictionary of arguments. This allows you to pass in named "key-value" pairs.

Use Case: When you want to store varying types of information about an object (like user profiles or configuration settings) without pre-defining every possible field.

The Name: Standard convention uses **kwargs** (Keyword Arguments).
