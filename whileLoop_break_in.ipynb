```python
# 1. Dataset (imagine this is imported from a file or API)
transactions = [
    {'id': 7001, 'amount': 150.00, 'status': 'success'},
    {'id': 7002, 'amount': 99.50, 'status': 'failed'},
    {'id': 7003, 'amount': 25.00, 'status': 'success'},
    {'id': 7004, 'amount': 150.00, 'status': 'refunded'},
    {'id': 7005, 'amount': 300.25, 'status': 'success'},
    {'id': 7006, 'amount': 45.10, 'status': 'success'},
]

# 2. Initialize a variable to hold the total revenue
total_revenue = 0.0

print("Processing sales data...\n")

# 3. Loop through each transaction
for transaction in transactions:
    # If the transaction was not successful, skip to the next one
    if transaction['status'] != 'success':
        print(f"--> Skipping transaction {transaction['id']} (Status: {transaction['status']})")
        continue

    # This part only runs for successful transactions
    print(f"✅ Processing successful transaction {transaction['id']}...")
    total_revenue += transaction['amount']


# 4. Print the final result
print(f"\nTotal revenue from successful sales: ${total_revenue:.2f}")
```
